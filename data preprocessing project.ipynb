```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/Users/rohanshenoy/Desktop/aero/data preprocessing/weatherAUS 2.csv')
data
#showing initial data
```

```python
#we drop the a few columns from the dataset and new dataset is X

label_encoder = LabelEncoder()
#we use label encoder to emcode the categorical column which have strings 
data['WindGustDir'] = label_encoder.fit_transform(data['WindGustDir'])
data['WindDir9am'] = label_encoder.fit_transform(data['WindDir9am'])
data['WindDir3pm'] = label_encoder.fit_transform(data['WindDir3pm'])
data['Location'] = label_encoder.fit_transform(data['Location'])
data
```

```python
data = data.fillna(0) #replacing nan values with zero
data 
```

```python
data.drop('Date',axis=1,inplace=True) #dropping date column as its useless/meaningless
data
```

```python
data.replace(['No', 'Yes'], [0,1], inplace=True)
data
#binary encoding
```

```python
#can also use alternative approach of one hot encoding

#performing one hot encoding using get_dummies fn on the columns having str value
X=pd.read_csv('/Users/rohanshenoy/Desktop/aero/data preprocessing/weatherAUS 2.csv')
X = pd.get_dummies(data, columns=['Location','RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
X
```

```python
data1=data #making a copy of the data so that i can show diff types of scaling
```

```python
#now that our data is a little better we can do feature scaling

features = data.drop(columns='RainTomorrow', axis=1)
Y = data['RainTomorrow']
```

```python
#performing min-max normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Convert the scaled features back to a DataFrame for easier handling
features_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)

# Display the scaled features
print("\nScaled Features:")
print(features_scaled_df.head())
```

```python
#performing standardization

features1 = data1.drop(columns='RainTomorrow', axis=1)
Y = data1['RainTomorrow']

from sklearn.preprocessing import StandardScaler

# Perform feature scaling using standardization
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features1)

# Convert the scaled features back to a DataFrame for easier handling
features_scaled_df = pd.DataFrame(features_scaled, columns=features1.columns)

# Display the scaled features
print("\nScaled Features:")
print(features_scaled_df.head())
```

```python
#dividing data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
```
